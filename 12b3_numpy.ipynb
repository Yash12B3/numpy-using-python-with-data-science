### 5.2 crop and flip operation using NumPy Indexing.

```python
# Crop only vertical portion, keep full width
row_crop = img[100:400, :]
plt.imshow(row_crop)
plt.axis('off')
```

```python
# Crop only horizontal portion, keep full height
col_crop = img[:, 150:450]
plt.imshow(col_crop)
plt.axis('off')
```

```python
#vertical flip from top to bottom
# Reverse rows using NumPy slicing
vertical_flip = img[::-1, :]
plt.imshow(vertical_flip)
plt.axis('off')
```

```python
# Reverse both rows and columns
both_flip = img[::-1, ::-1]
plt.imshow(both_flip)
plt.axis('off')
```

```python
# Horizontal flip
# Vertical flip
flip_h = np.flip(img, axis=1)
flip_v = np.flip(img, axis=0)
plt.imshow(flip_h)
plt.axis('off')
```

```python
# Crop center dynamically
h, w = img.shape[:2]
center_crop = img[h//4:3*h//4, w//4:3*w//4]
plt.imshow(center_crop)
plt.axis('off')
```

```python
# Calculate middle column
mid = w // 2
# Left half
left_half = img[:, :mid]
# Right half
right_half = img[:, mid:]
# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(left_half)
plt.title("Left Half")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(right_half)
plt.title("Right Half")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Number of vertical parts
n = 5   # change this to any number
# Get image dimensions
h, w, c = img.shape
part_width = w // n
plt.figure(figsize=(12, 4))
for i in range(n):
    # Slice columns dynamically
    start = i * part_width
    end = (i + 1) * part_width if i != n - 1 else w
    vertical_part = img[:, start:end]
    plt.subplot(1, n, i + 1)
    plt.imshow(vertical_part)
    plt.title(f"Part {i + 1}")
    plt.axis('off')
plt.show()
```

```python
#converting photos into shuffle pieces
import numpy as np
import matplotlib.pyplot as plt
# number of puzzle rows and columns
rows = 4
cols = 4
# get image height and width
h, w = img.shape[:2]
h_part = h // rows
w_part = w // cols
pieces = []
# split image into pieces
for i in range(rows):
    for j in range(cols):
        r1 = i * h_part
        r2 = (i + 1) * h_part if i != rows - 1 else h
        c1 = j * w_part
        c2 = (j + 1) * w_part if j != cols - 1 else w
        pieces.append(img[r1:r2, c1:c2])
# shuffle pieces
np.random.shuffle(pieces)
# display puzzle
plt.figure(figsize=(8, 8))
for i in range(len(pieces)):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(pieces[i])
    plt.axis('off')
plt.show()
```
